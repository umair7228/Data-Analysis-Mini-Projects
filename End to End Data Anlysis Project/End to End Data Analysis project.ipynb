```python
import pandas as pd
```

```python
df = pd.read_csv('./orders.csv')
df.head(10)
```

```python
# As we can see that there are some not_availabe and unknown values in the Ship Mode column and we need to change it to null
# First let see the the all uniqe values in the Ship Mode column
df['Ship Mode'].unique()
```

```python
# handling null values
# There are Not available and unknown  values in the column and we need to change it to null
df = pd.read_csv('./orders.csv', na_values=['Not Available', 'unknown'])
df['Ship Mode'].unique()
```

```python
# Now as we can see the 'Not Available', 'unknown' values has been changed in to nan

# rename the columns names, making them in lower case and replace space with underscore

# lets convert column names in lower case
df.columns=df.columns.str.lower()
df.columns

# now replaceing the space with underscore
df.columns=df.columns.str.replace(' ', '_')
df.head()
```

```python
# Now we have the discount column in percentage so we will derive new columns of actual value of discrount, sale price and profit
# lets find the discount 
df['discount'] = df['list_price'] * df['discount_percent'] * 0.1
df.head(3)
```

```python
# now we have discount for each product so we can calculate the sale price
df['sale_price'] = df['list_price'] - df['discount']
df.head(2)
```

```python
# Now we have also the sale price so we could calculate profit by subtracting cost_price from sale_price
df['profit'] = df['cost_price'] - df['sale_price']
df.head(2)
```

```python
df.dtypes
```

```python
# If we see the data type of order_date, it is in object and we need to convert it to date data type
df['order_date'] = pd.to_datetime(df['order_date'], format="%Y-%m-%d")
df.dtypes
```

```python
# Now we will drop the cost_price, list_price and discount_percent columns because these columns are non beneficial for us now
df.drop(columns=['cost_price', 'list_price', 'discount_percent'], inplace=True)
```
